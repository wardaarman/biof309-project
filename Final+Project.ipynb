#### Objective: To visualize microarray data; specifically as a volcano plot


##### import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.ticker as ticker
```

##### importing the dataset:
#### Download Supplemental Data Table II from https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2528102/bin/pp.108.121038_index.html
#### Just make sure to save the file as a ‘csv’ file and note the location

```python
df3 = pd.read_csv('C:/Users/wardaarman/Desktop/Supplemental Table 2.csv', sep=',', usecols=[1,2, 3])

##check that import works with .head() attribute

print(df3.head())
```

```python
#Name variables for values plotted along x-axis and those plotted along the y-axis based on column
#header names

FC= df3['Log2FC']
pvalue= df3['p-value']

#Set axes tick marks by using 'ax=plt.axes(). Then set major tick marks and/or minor tick marks along either axis.
#The x-axis is specifically changed here.

ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.75))

#Call the plot function with the scatter attribute. Here use the parameters to plot x by y and decide on color, alpha, etc.

plt.scatter(FC, pvalue, c= 'purple', alpha= 0.25)

#Print the max and min values from the columns selected for the used data frame to determine limits on x and y axes.
print(np.max(df3))
print(np.min(df3))

#Set limits on x and y axes

plt.xlim(-4,4)
plt.ylim(0.001, 0 )

#Change title and x and y labels on plot.

plt.title('Gene expression changes in Arabidopsis infected with CaLCuV', fontsize = 10)
plt.ylabel('p-value', fontsize = 10)
plt.xlabel('Fold change', fontsize = 10)

#Show the plot.
plt.show()
```
